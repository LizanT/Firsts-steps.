# Simple Linear Regression.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
salary = pd.read_csv('salary_clean.csv')
```

```python
salary
```

# Salary in Function Years of Experience.

```python
plt.figure(figsize=(10,6))

plt.ylabel('Salary ($)')
plt.xlabel('Years of Experience')
plt.scatter(salary["Years of Experience"], salary["Salary"], color="pink")
plt.show()
```

# Model Creaction.

```python
from sklearn import linear_model

regression = linear_model.LinearRegression()

# Changes into vector.
years = salary['Years of Experience'].values.reshape((-1,1))
```

# Training the model.

```python
# Using fit, the model will train with the input and ouput values.
model = regression.fit(years, salary['Salary'])

print(f'Intercept (b) {model.intercept_}')
print(f'Slope (m) {model.coef_} ')
```

# Prediction with a new vector.

```python
# People who has 7 , 13, 15, 21 years of experience, their salary will be...

values = [[5],[13],[15],[30]]

predict_values = model.predict(values)

print(f'A person who has 7, 13, 15 or 21 years of experience will be earn : {predict_values}')
```

# Plot prediction into dataframe chart.

```python
plt.figure(figsize=(10,6))

plt.scatter(values, predict_values, color='green', s=50, zorder=10)
plt.plot(values, predict_values, color='black')

# Dataframe chart
plt.ylabel('Salary ($)')
plt.xlabel('Years of Experience')
plt.scatter(salary["Years of Experience"], salary["Salary"], color="pink")
plt.show()
```
