## Preguntas clave.
#### ¿Qué producto genera más ingresos?
#### ¿En qué mes se vendió más cada producto?
#### ¿Cómo es la tendencia de ventas a lo largo del año?

```python
#Intalamos las librerias.
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

## Creación de datos con Numpy.

```python
#Configuramos la semilla que dará los datos aleatorios.
np.random.seed(42)
```

```python
#Creamos el array donde introduciremos los números; estos representarán la venta de cuatro productos en doce meses.
ventas = np.random.randint(50,500, size=(12,4))
```

## Estructurar datos con Pandas.

```python
#Definimos las filas y las columnas.
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
productos = ['Smartphones', 'Ordenadores', 'Tablets', 'Laptops']
```

```python
#Creamos el DataFrame.
df_ventas = pd.DataFrame(ventas, index=meses, columns=productos) 

#Donde 'pd.Dataframe' convierte el array de Numpy en una tabla (ventas, meses, productos) de Pandas (o matriz); posteriormente se crea el DataFrame.
#'index' nombra las filas y 'columns' las columnas de la tabla.
```

```python
df_ventas
```

## Análisis de ventas.

```python
#Definimos los precios de nuestros cuatro productos.
precios = np.array([800,1500,300,600]) #Este array, como podemos observar es un vector.
```

```python
#Calculamos los ingresos (ventas x precios).
ingresos = df_ventas * precios

#Aquí estamos multiplicando nuestro DataFrame por el array.
#Podemos hacer un símil de multipliación entre una matriz y un vector.
```

```python
#Aplicamos Estadística descriptiva a nuestros datos (media,desviación estándar, percentil etc).
estadisticas = df_ventas.describe() 

#describe() genera estadísticas como la media, desviación estándar etc.
```

```python
estadisticas
```

## Visualización con Matplotlib.

```python
#Definimos el tamaño del gráfico.
plt.figure(figsize=(20,5))

#Creamos un diccionario para asignar a cada producto un color.
colores_productos = {
    'Smartphones': 'red',
    'Ordenadores': 'blue',
    'Tablets': 'forestgreen',
    'Laptops': 'darkorange'
}

#Creamos un bucle para que cada iteración tome un valor diferente de cada producto.
for producto in productos:
    plt.plot(
        meses,                                  #Marca el eje X.
        df_ventas[producto],                    #Marca el eje Y.
        marker='o',                             #Forma del punto.
        linestyle='-',                          #Estilo de línea, tenemos opciones como: ('--'),(';'),('-.').
        color=colores_productos[producto],      #Color de la línea de cada producto.
        linewidth=2,                            #Grosor de línea.
        markersize=8,                           #Tamaño del marcador.
        label=producto,                         #Nombre de la leyenda (Smartphones,Ordenadores,Tablets,Laptops).
    )
plt.title('Evolución de Ventas Mensuales por Producto.')
plt.xlabel('Meses.')
plt.ylabel('Unidades vendidas.')
plt.legend()
plt.grid(True)
plt.show
```

## Análisis de los datos.

```python
#Producto con más ingresos totales.
ingresos_totales = ingresos.sum()
producto_top = ingresos_totales.idxmax() #Es entonces que al usar 'idxmax' conoceremos el producto más vendido.
```

```python
print('Los ingresos totales en el año es:')
print(ingresos_totales)
print('El producto más vendido es:',producto_top)
```

```python
#Mes con más ventas totales.
meses_total = df_ventas.sum(axis=1)
mes_top = meses_total.idxmax
```

```python
print('El mes con más ventas fue',mes_top)
```
